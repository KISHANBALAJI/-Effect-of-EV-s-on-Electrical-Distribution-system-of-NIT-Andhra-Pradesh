```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('Copy of MINIPROJECT_random.final(random)(1).csv')
X_0 = dataset.iloc[:,0].values
y_0 = dataset.iloc[:,2].values
y_1 = dataset.iloc[:,4].values
y_2 = dataset.iloc[:,6].values
y_3 = dataset.iloc[:,8].values
```

```
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
X_0=X_0.reshape(-1,1)
regressor.fit(X_0, y_0)
y_0_pred=regressor.predict(X_0)
```

```
X_0_grid = np.arange(min(X_0), max(X_0), 0.01)
X_0_grid = X_0_grid.reshape((len(X_0_grid), 1))
#plt.scatter(X, y, color = 'red')
plt.plot(X_0_grid, regressor.predict(X_0_grid), color = 'blue')
plt.title("Load(KWh) vs Time(hrs) - Case_0")
plt.ylabel('Load(KWh)')
plt.xlabel('Time(hrs)')
plt.show()
```
